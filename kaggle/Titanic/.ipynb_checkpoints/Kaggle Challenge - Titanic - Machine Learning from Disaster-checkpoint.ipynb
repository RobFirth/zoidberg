```python
%matplotlib notebook

import matplotlib.pyplot as plt
import matplotlib.cm as cmap

import numpy as np
import pandas as pd

import seaborn as sb
```

```python
seed = 54
```

```python
train_df = pd.read_csv("../data/train.csv")
test_df = pd.read_csv("../data/test.csv")
```

```python
test_df.head()
```

```python
train_df.head()
```

```python
train_df.describe()
```

```python
wanted_columns = ["Age", "Fare", "Sex", "SibSp"]
plt.figure(figsize=(10, 4))

for index, wanted_column in enumerate(wanted_columns):
    plt.subplot(1, 4, index + 1)
    sb.violinplot(x="Survived", y=wanted_column, data=train_df[wanted_columns+["Survived"]])
```

```python
def gender_bool(gender):
    if gender.lower() == "male":
        return True
    if gender.lower() == "female":
        return False
    else:
        return np.nan
    
def numerate(boolean):
    return int(boolean)
```

```python
train_df.loc[:,"gender_bool"] = train_df.Sex.map(gender_bool).map(numerate)
test_df.loc[:,"gender_bool"] = test_df.Sex.map(gender_bool).map(numerate)
```

```python
print(train_df.shape)
print(test_df.shape)
```

```python
wanted_columns_train = ["Age", "Fare", "gender_bool", "SibSp", "Survived"]
wanted_columns_test = ["Age", "Fare", "gender_bool", "SibSp"]
# df = df.drop(df.loc[df["Gun Time"] == "DNF"].index)
cut_train_df = train_df[wanted_columns_train].dropna()
cut_test_df = test_df[wanted_columns_test].dropna()

print(cut_train_df.shape)
print(cut_test_df.shape)
```

```python
X = cut_train_df[wanted_columns_test]
y = cut_train_df["Survived"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=seed)
```

```python
# Step 1: Import the model
from sklearn.tree import DecisionTreeClassifier

# Step 2: Instantiate the model
decision_tree_classifier = DecisionTreeClassifier(random_state=seed)

# Step 3: Fit the model on data (i.e. train the model)
decision_tree_classifier.fit(X_train, y_train)

# Step 4: Generate predictions / scores
decision_tree_classifier.score(X_test, y_test)
```

```python
model_accuracies = []
plt.figure(figsize=(8, 4))

# Split the dataset differently and fit a model on this split, 1,000 times
for i in range(1000):
    X2_train, X2_test, y2_train, y2_test = train_test_split(X, y, train_size=0.8, random_state=i)
    
    decision_tree_classifier = DecisionTreeClassifier(random_state=seed)
    decision_tree_classifier.fit(X2_train, y2_train)
    classifier_accuracy = decision_tree_classifier.score(X2_test, y2_test)
    model_accuracies.append(classifier_accuracy)

sb.distplot(model_accuracies)
```

```python
from sklearn.cross_validation import StratifiedKFold

def plot_cv(cv, n_samples):
    masks = []
    for train, test in cv:
        mask = np.zeros(n_samples, dtype=bool)
        mask[test] = 1
        masks.append(mask)
        
    plt.figure(figsize=(10, 2))
    plt.imshow(masks, interpolation='none')
    plt.ylabel('Fold')
    plt.xlabel('Row #')

# Use a subset of 100 examples
subset = y[0:100]
    
plot_cv(StratifiedKFold(subset, n_folds=10), len(subset))
```

```python
from sklearn.model_selection import cross_val_score

decision_tree_classifier = DecisionTreeClassifier(random_state=seed)

# cross_val_score returns a list of the scores, which we can visualize
# to get a reasonable estimate of our classifier's performance
cv_scores = cross_val_score(decision_tree_classifier, X_train, y_train, cv=10)
plt.figure(figsize=(8, 4))
sb.distplot(cv_scores)
plt.title('Average score: {}'.format(np.mean(cv_scores)))
```

```python
# Step 1: Import the model
from sklearn.svm import SVC

# Step 2: Instantiate the model
svclassifier = SVC(random_state=seed)

# Step 3: Fit the model on data (i.e. train the model)
svclassifier.fit(X_train, y_train)

# Step 4: Generate predictions / scores
svclassifier.score(X_test, y_test)
```

```python
model_accuracies = []
plt.figure(figsize=(8, 4))

# Split the dataset differently and fit a model on this split, 1,000 times
for i in range(1000):
    X2_train, X2_test, y2_train, y2_test = train_test_split(X, y, train_size=0.8, random_state=i)
    
    svclassifier = SVC(random_state=seed)
    svclassifier.fit(X_train, y_train)
    classifier_accuracy = svclassifier.score(X_test, y_test)
    model_accuracies.append(classifier_accuracy)

sb.distplot(model_accuracies)
```

```python
def is_married(name):
    if "mrs" in name.lower():
#         print("Married Woman")
        return 1
    if "miss" in name.lower():
        return 2
    else:
        return 0
```

```python
def embarked(letter):
    if type(letter) == str:
        if letter.lower() == "q":
            return 1
        if letter.lower() == "s":
            return 2
        if letter.lower() == "c":
            return 3
        else:
            return -1
    else:
        return -1
```

```python
train_df.loc[:,"married_num"] = train_df.Name.map(is_married)
test_df.loc[:,"married_num"] = test_df.Name.map(is_married)

train_df.loc[:,"embarked_num"] = train_df.Embarked.map(embarked)
test_df.loc[:,"embarked_num"] = test_df.Embarked.map(embarked)
```

```python
train_df
```

```python
wanted_columns_train = ["Age", "Fare", "gender_bool", "SibSp", "Survived", "married_num", "embarked_num"]
wanted_columns_test = ["Age", "Fare", "gender_bool", "SibSp", "married_num", "embarked_num"]
# df = df.drop(df.loc[df["Gun Time"] == "DNF"].index)
cut_train_df = train_df[wanted_columns_train].dropna()
cut_test_df = test_df[wanted_columns_test].dropna()

print(cut_train_df.shape)
print(cut_test_df.shape)
```

```python
X = cut_train_df[wanted_columns_test]
y = cut_train_df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=seed)
```

```python
# Step 1: Import the model
from sklearn.tree import DecisionTreeClassifier

# Step 2: Instantiate the model
decision_tree_classifier = DecisionTreeClassifier(random_state=seed)

# Step 3: Fit the model on data (i.e. train the model)
decision_tree_classifier.fit(X_train, y_train)

# Step 4: Generate predictions / scores
decision_tree_classifier.score(X_test, y_test)
```

```python
X.describe()
```

```python
model_accuracies = []
plt.figure(figsize=(8, 4))

# Split the dataset differently and fit a model on this split, 1,000 times
for i in range(1000):
    X2_train, X2_test, y2_train, y2_test = train_test_split(X, y, train_size=0.8, random_state=i)
    
    decision_tree_classifier = DecisionTreeClassifier(random_state=seed)
    decision_tree_classifier.fit(X2_train, y2_train)
    classifier_accuracy = decision_tree_classifier.score(X2_test, y2_test)
    model_accuracies.append(classifier_accuracy)

sb.distplot(model_accuracies)
```

```python
from sklearn.model_selection import GridSearchCV

decision_tree_classifier = DecisionTreeClassifier(random_state=seed)

parameter_grid = {'max_depth': [1, 2, 3, 4, 5, 6],
                  'max_features': [1, 2, 3, 4, 5, 6]}

cross_validation = StratifiedKFold(y_train, n_folds=10, random_state=seed)

grid_search = GridSearchCV(decision_tree_classifier,
                           param_grid=parameter_grid,
                           cv=cross_validation)

grid_search.fit(X_train, y_train)
print('Best score: {}'.format(grid_search.best_score_))
print('Best parameters: {}'.format(grid_search.best_params_))

# Get the best estimator from the grid search
decision_tree_classifier = grid_search.best_estimator_

# Score the estimator on the test dataset
decision_tree_classifier.score(X_test, y_test)
```

```python
# Step 1: Import the class
from sklearn.ensemble import RandomForestClassifier

# Step 2: Instantiate the estimator
random_forest_classifier = RandomForestClassifier(random_state=seed)

# Hyperparameter values to explore
parameter_grid = {'n_estimators': [5, 10, 25, 50],
                  'criterion': ['gini', 'entropy'],
                  'max_features': [1, 2, 3, 4, 5, 6],
                  'warm_start': [True, False]}

# Create a cross validation object using StratifiedKFold
cross_validation = StratifiedKFold(y_train, n_folds=10, random_state=seed)

# Instantiate the grid search using GridSearchCV 
grid_search = GridSearchCV(random_forest_classifier,
                           param_grid=parameter_grid,
                           cv=cross_validation)

# Step 3 & 4: Fit the estimator on data (i.e. train the model) and generate predictions
grid_search.fit(X_train, y_train)

# Display the best model scores and parameters
print('Best score: {}'.format(grid_search.best_score_))
print('Best parameters: {}'.format(grid_search.best_params_))

grid_search.best_estimator_
```

```python
# Get the best estimator from the grid search
random_forest_classifier = grid_search.best_estimator_

# Score the estimator on the test dataset
random_forest_classifier.score(X_test, y_test)
```

```python
import string

def cabin_class(cabin):
    if type(cabin) is str:
        #         return cabin[0]
        return string.ascii_lowercase.index(cabin[0].lower())
    elif np.isnan(cabin):
        return -1
    pass
```

```python
train_df.loc[:,"cabin_flag"] = train_df.Cabin.map(cabin_class)
test_df.loc[:,"cabin_flag"] = test_df.Cabin.map(cabin_class)
```

```python
X = cut_train_df[wanted_columns_test]
y = cut_train_df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=seed)
```

```python
# Step 1: Import the model
from sklearn.tree import DecisionTreeClassifier

# Step 2: Instantiate the model
decision_tree_classifier = DecisionTreeClassifier(random_state=seed)

# Step 3: Fit the model on data (i.e. train the model)
decision_tree_classifier.fit(X_train, y_train)

# Step 4: Generate predictions / scores
decision_tree_classifier.score(X_test, y_test)
```
